# **Feature Extraction using VGG16:**

```python
model_vgg16 = VGG16()
model_vgg16 = Model(inputs=model_vgg16.inputs,outputs=model_vgg16.layers[-2].output)
model_vgg16.summary()
```

```python
features_vgg16 = {}
dirc = "Round_1_AI/"
for img in os.listdir(dirc):
  img_path = dirc+img
  image = load_img(img_path,target_size=(224,224))
  image = img_to_array(image)
  image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
  image = preprocess_input(image)
  feature_vgg = model_vgg16.predict(image)
  features_vgg16[img] = feature_vgg
```

# **Feature Extraction using DenseNet:**

```python
from tensorflow.keras.applications.densenet import DenseNet121
```

```python
model_dense = DenseNet121(weights='imagenet', include_top=False, pooling='avg')
model_dense.summary()
```

```python
features_Dense = {}
dirc = "Round_1_AI/"
for img in os.listdir(dirc):
  img_path = dirc+img
  image = load_img(img_path,target_size=(224,224))
  image = img_to_array(image)
  image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
  image = preprocess_input(image)
  feature_Dense = model_dense.predict(image)
  features_Dense[img] = feature_Dense
```
